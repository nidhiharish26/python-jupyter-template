### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
# Data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

if len(x) != len(y):
    raise ValueError("x and y lists must have same length")
# Create a line plot
plt.plot(x, y)

# Labeling the axes and the plot
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Basic Line Plot Graph')

plt.grid(True)

# Show the plot
plt.show()
```
