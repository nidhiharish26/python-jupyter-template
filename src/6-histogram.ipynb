### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
# Data
import matplotlib.pyplot as plt
import numpy as np
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

if not isinstance(data, (list, tuple, np.ndarray)):  # Add np.ndarray if you use NumPy
    raise TypeError("Data must be a list, tuple, or numpy array.")

if not data:  # or if len(data) == 0:
    raise ValueError("Data cannot be empty.")

if not all(isinstance(x, (int, float)) for x in data):
    raise TypeError("All elements in data must be numeric (int or float).")

# Create a histogram
plt.hist(data, bins=5, color = 'red')

# Labeling the axes and the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Histogram Analysis')

# Show the plot
plt.show()
```
